```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from pandas import Series
from datetime import datetime
```

```python
train=pd.read_csv('ts_train.csv')
test=pd.read_csv('ts_test.csv')
```

```python
# standard practice to copy data
train_orig= train.copy()
test_orig=test.copy()
```

```python
#feature extraction 

#converting object to datetime (Avoid this to run multiple times.. Just run it one time)
train['Datetime']=pd.to_datetime(train.Datetime, format='%d-%m-%Y %H:%M')
test['Datetime']=pd.to_datetime(test.Datetime, format='%d-%m-%Y %H:%M')
train_orig['Datetime']=pd.to_datetime(train.Datetime, format='%d-%m-%Y %H:%M')
test_orig['Datetime']=pd.to_datetime(test.Datetime, format='%d-%m-%Y %H:%M')
```

```python
train.dtypes, test.dtypes
```

```python
# Adding columns in data set

for i in (train, test, train_orig, test_orig):
    i['year']=i.Datetime.dt.year
    i['month']=i.Datetime.dt.month
    i['day']=i.Datetime.dt.day
    i['hour']=i.Datetime.dt.hour
```

```python
train['day of week']= train['Datetime'].dt.dayofweek
temp= train['Datetime']
```

```python
def checkweekend(data):
    if data.dayofweek== 5 or data.dayofweek==6:
        return 1
    else:
        return 0
t2= train['Datetime'].apply(checkweekend)
train['weekend']=t2
```

```python
#plotting data on diff axes

train.index= train['Datetime']

# creating seperate variable for plotting (dropping ID)
df=train.drop('ID',1)
ts=df['Count']
plt.figure(figsize=(16,8))
plt.plot(ts, label="Passenger count")
plt.title("Time Series")
Plt.xlabel('Time(year-month)')
plt.ylabel('Passenger count')
plt.legend(loc='best')
```

```python
# Resampling- TO reduce the noise
train['Timestamp']= pd.to_datetime(train.Datetime, format= '%d-%m-%Y-%H:%M')
train.index= train.Timestamp
hourly= train.resample('H').mean()
daily= train.resample('D').mean()
weekly= train.resample('W').mean()
monthly= train.resample('M').mean()
```

```python
fig, axs=plt.subplots(4,1)

hourly.Count.plot(figsize=(15,8), title='Hourly', fontsize=14, ax=axs[0])
daily.Count.plot(figsize=(15,8), title='Daily', fontsize=14, ax=axs[0])
weekly.Count.plot(figsize=(15,8), title='Monthly', fontsize=14, ax=axs[0])
monthly.Count.plot(figsize=(15,8), title='Yearly', fontsize=14, ax=axs[0])

plt.show()
```
